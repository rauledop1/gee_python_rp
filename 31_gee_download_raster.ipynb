```python
import ee
import requests
from google.oauth2 import service_account

# Authenticate and initialize the Earth Engine API using the service account
service_account_credentials = 'path_to_your_service_account_credentials.json'
credentials = service_account.Credentials.from_service_account_file(service_account_credentials)
scoped_credentials = credentials.with_scopes(['https://www.googleapis.com/auth/earthengine.readonly'])

# Initialize Earth Engine with the credentials
ee.Initialize(credentials=scoped_credentials)

# Define the region of interest (ROI)
roi = ee.Geometry.Rectangle([-122.275, 37.871, -122.259, 37.885])

# Select the image (Landsat 8 image in this example)
image = ee.Image('LANDSAT/LC08/C01/T1_SR/LC08_044034_20200820').clip(roi)

# Get the download URL
url = image.getDownloadURL({
    'scale': 30,  # Resolution in meters
    'region': roi
})

print('Download URL:', url)

# Download the image using the token
headers = {
    'Authorization': 'Bearer ' + scoped_credentials.token
}
response = requests.get(url, headers=headers)

# Save the image to a file
with open('align.tif', 'wb') as file:
    file.write(response.content)

print('Image downloaded successfully.')
```
